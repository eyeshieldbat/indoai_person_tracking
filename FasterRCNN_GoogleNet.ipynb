```python
def xywh_to_xyxy(bbox_tensor):
  bbox_xyxy_tensor =  bbox_tensor.clone()
  bbox_xyxy_tensor[:,2] +=bbox_tensor[:,0] #xmax = x + w
  bbox_xyxy_tensor[:,3] +=bbox_tensor[:,1]  #ymax = x + h

  return bbox_xyxy_tensor
```

```python
class ObjectDataset(torch.utils.data.Dataset):

  def __init__(self, coco_dataset, transforms=None):
    # please define the data proses init
    self.coco_dataset = coco_dataset
    self.transforms= transforms
    self.person_categori_id = 1 
    
    self.ids= [sample.id for sample in coco_dataset]
    

  def __getitem__(self, idx):
    # please define the dataloader
    sample_id = self.ids[idx]
    sample = self.coco_dataset[sample_id]
    #load image
    img = Image.open(sample.filepath).convert("RGB")
   
    # convert everything into a torch.Tensor
    boxes = torch.as_tensor([ann.bounding_box for ann in sample.ground_truth.detections], dtype=torch.float32)

    boxes[:,0] *= img.size[0] #x1*width
    boxes[:,1] *= img.size[1] #y1*height
    boxes[:,2] *= img.size[0] #x2*width
    boxes[:,3] *= img.size[1] #y2*height



    target = {}
    target['boxes'] = xywh_to_xyxy(boxes)
    target['labels'] = torch.tensor([self.person_categori_id for ann in sample.ground_truth.detections], dtype=torch.int64)
    target['image_id'] = torch.tensor([idx])
    target['area'] = boxes[:,3] * boxes[:,2]
    target['iscrowd'] = torch.tensor([ann.iscrowd for ann in sample.ground_truth.detections], dtype=torch.int64)

    if self.transforms is not None: # preprocessing and augmentation
      img, target = self.transforms(img, target)

    return img, target
  
  def __len__(self):
    return len(self.ids)
```

```python
import transforms as T

def get_transform(train):
    # fill the data transformation techniques here
    transforms = []
    transforms.append(T.PILToTensor())
    transforms.append(T.ConvertImageDtype(torch.float))
    if train:
        transforms.append(T.RandomHorizontalFlip(0.5))
    
    return T.Compose(transforms)
```

```python
# use our dataset and defined transformations
train_data = ObjectDataset(dataset, get_transform(train=True))
test_data = ObjectDataset(dataset_test, get_transform(train=False))


# define training and validation data loaders
train_data_loader = torch.utils.data.DataLoader(
    train_data, batch_size=2, shuffle=True, num_workers=2,
    collate_fn=utils.collate_fn)
  
test_data_loader_test = torch.utils.data.DataLoader(
    test_data, batch_size=1, shuffle=False, num_workers=2,
    collate_fn=utils.collate_fn)
```

```python
import torch
import torchvision
from torchvision.models.detection import FasterRCNN
from torchvision.models.detection.rpn import AnchorGenerator

# backbone googlenet
googlenet = torchvision.models.googlenet(weights="DEFAULT")

backbone = torch.nn.Sequential(*list(googlenet.children())[:-3]) # menghilangkan fully connected layer dan global avg pool

backbone.out_channels = 1024 # Define the anchor generator

anchor_generator = AnchorGenerator(sizes=((32, 64, 128, 256, 512),),
                                   aspect_ratios=((0.5, 1.0, 2.0),))

# Define the ROI pooler
roi_pooler = torchvision.ops.MultiScaleRoIAlign(featmap_names=['0'],
                                                output_size=7,
                                                sampling_ratio=2)

# Create the Faster R-CNN model with GoogleNet backbone
model = FasterRCNN(backbone, num_classes=2, rpn_anchor_generator=anchor_generator, box_roi_pool=roi_pooler)
```

```python
def train_epoch(model, dataloader, optimizer, device, epoch):
  # do training and calculate training loss
  train_one_epoch(model, optimizer, dataloader, device, epoch, print_freq=10)

def evaluate_model(model, dataloader, device):
  # calculate mAP for the data evaluation
  evaluate(model, dataloader, device=device)
```

```python
def train(model, train_dataloader, test_dataloader, num_epochs):
    # train on the GPU or on the CPU, if a GPU is not available
    device = torch.device('cuda') if torch.cuda.is_available() else torch.device('cpu')
    # device = torch.device('cpu')
    # move model to the right device
    model.to(device)

    # construct an optimizer
    params = [p for p in model.parameters() if p.requires_grad]
    optimizer = torch.optim.SGD(params, lr=0.005, momentum=0.9, weight_decay=0.0005)
    # and a learning rate scheduler
    lr_scheduler = torch.optim.lr_scheduler.StepLR(optimizer, step_size=3, gamma=0.1)
    
    for epoch in range(num_epochs):
        # train for one epoch, printing every 100 iterations
        train_one_epoch(model, optimizer, train_dataloader, device, epoch, print_freq=10)

        # update the learning rate
        lr_scheduler.step()
        
        # evaluate on the test dataset
        evaluate_model(model, test_dataloader, device=device)
```

```python
train(model, train_data_loader, test_data_loader_test, 10)
```

```python
# Choose a random subset of 100 samples to add predictions to
predictions_view = dataset_test.take(size=100,seed=51)

print(predictions_view)
```

```python

from torchvision.transforms import functional as func
# Get class list
classes = dataset.default_classes

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

# Add predictions to samples
with fo.ProgressBar() as pb:
    for sample in pb(predictions_view):
        # Load image
        image = Image.open(sample.filepath)
        image = func.to_tensor(image).to(device)
        c, h, w = image.shape

        # Perform inference
        preds = model([image])[0]
        labels = preds["labels"].cpu().detach().numpy()
        scores = preds["scores"].cpu().detach().numpy()
        boxes = preds["boxes"].cpu().detach().numpy()

        # Convert detections to FiftyOne format
        detections = []
        for label, score, box in zip(labels, scores, boxes):
            # Convert to [top-left-x, top-left-y, width, height]
            # in relative coordinates in [0, 1] x [0, 1]
            x1, y1, x2, y2 = box
            rel_box = [x1 / w, y1 / h, (x2 - x1) / w, (y2 - y1) / h]

            detections.append(
                fo.Detection(
                    label=classes[label],
                    bounding_box=rel_box,
                    confidence=score
                )
            )

        # Save predictions to dataset
        sample["faster_rcnn"] = fo.Detections(detections=detections)
        sample.save()

print("Finished adding predictions")
```

```python
from fiftyone import ViewField as F
# Only contains detections with confidence >= 0.75
high_conf_view = predictions_view.filter_labels("faster_rcnn", F("confidence") > 0.75, only_matches=False)
```

```python
# Evaluate the predictions in the `faster_rcnn` field of our `high_conf_view`
# with respect to the objects in the `ground_truth` field
results = high_conf_view.evaluate_detections(
    "faster_rcnn",
    gt_field="ground_truth",
    eval_key="eval",
    compute_mAP=True,
)
```

```python
#load data test
data_test = ObjectDataset(dataset_test, None)
def preprocess(image, transform):
  return transform(image, None) # image, target

def inference(model, x, device):
  checkpoint = torch.load('/content/drive/MyDrive/save model/modelfasterrcnngooglenet.pt')
  model.load_state_dict(checkpoint)
  model.eval()
  x = torch.unsqueeze(x, 0) #Neural Network B, C, H, W

  x = x.to(device)

  x = model(x)

  return x

def postprocess(output, thr=0.5): #filtering terhadap score, jika score threshold dibawah 0.5 maka ouput tidak digunakan dan jika score thresholdnya diatas 0.5 maka digunakan
  # 1. aku melakukan concat terhadap bboxes, labels, scores menjadi satu buah baris tensor
  # [x1,y1,x2,y2, 1, scores]
  output = output[0]
  boxes = output['boxes']
  labels = output['labels'].unsqueeze(1)
  scores = output['scores'].unsqueeze(1)

  converted_data = torch.cat((boxes, labels, scores), dim=1)

  filtered_data = converted_data[converted_data[:, 5] > thr]

  return filtered_data

def visualize(image, filtered_data):
    # Scale the filtered_data bounding boxes
    height, width, _ = image.shape
    scaled_data = filtered_data.clone().detach().cpu()

    # create a cpoy of the image to draw on
    image_with_boxes = image.copy()

    #Draw bounding boxes and labels
    for bbox in scaled_data:
        xmin, ymin, xmax, ymax = bbox[:4].tolist()
        label = "Person"
        #Draw the rectangle on the image
        cv2.rectangle(image_with_boxes, (int(xmin), int(ymin), 
                                        int(xmax), int(ymax)), 
                                          (0, 255, 0), 2)

        #Add label text
        cv2.putText(image_with_boxes, label, (int(xmin), int(ymin)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
    
    # Display the image with bounding boxes
    cv2_imshow(image_with_boxes)
```

```python
image = data_test[3][0]
image_array = np.array(image)
image_array = cv2.cvtColor(image_array, cv2.COLOR_RGB2BGR)
image_tensor,_ = preprocess(image, get_transform(False))

output = inference(model, image_tensor, "cuda")
print(output)
output = postprocess(output, 0.5)

visualize(image_array, output)
```
